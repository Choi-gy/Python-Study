```python
#제주도의 남녀 성비 비교
import csv
f = open('gender4.csv')
data = csv.reader(f)

m=[]
f=[]

name = input('찾고 싶은 지역의 이름을 알려주세요 : ')
for row in data :
    if name in row[0] :
        for i in row[3:104] :
            m.append(-int(i.replace(',', '')))
        for i in row[106:] :
            f.append(int(i.replace(',', '')))
        break                                                  #row 데이터에 '제주특별자치도'의 값은 제일 위에 행에 있음, 그것만 가져오는것
        
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize=(10, 5), dpi=300)
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(name + ' 지역의 남녀 성별 인구 분포')
plt.barh(range(101), m, label='남성')
plt.barh(range(101), f, label='여성')
plt.legend()
plt.show()
```

```python
#제주도의 성별 인구 비율
##나이별 인원이 아닌 합계 데이터를 구해야함
import csv
f = open('gender4.csv')
data = csv.reader(f)
size = []                                               #새로운 변수 지정
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')

for row in data :
    if name in row[0] :
        m = 0
        f = 0
        for i in range(101) :                          #인덱스를 다루기 위함(0~100까지의 구간 지정)
            m += int(row[i+3].replace(',', ''))        #3~103 인덱스 더하기
            f += int(row[i+106].replace(',', ''))      #106~ 206 인덱스 더하기
            
        break

size.append(m)
size.append(f)
print(size)
```

```python
##파이 차트로 그리기
import csv
f = open('gender4.csv')
data = csv.reader(f)
size = []                                               #새로운 변수 지정
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')

for row in data :
    if name in row[0] :
        m = 0
        f = 0
        for i in range(101) :                          #인덱스를 다루기 위함(0~100까지의 구간 지정)
            m += int(row[i+3].replace(',', ''))        #3~103 인덱스 더하기
            f += int(row[i+106].replace(',', ''))      #106~ 206 인덱스 더하기
            
        break

size.append(m)
size.append(f)
print(size)

import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
color = ['crimson', 'darkcyan']
plt.axis('equal')
plt.pie(size, labels=['남', '여'], autopct='%.1f%%', colors=color, 
        startangle=90)  #파이 차트는 3시에서 반시계 방향으로 형성,
                        #startangle =90 으로 설정 시 , 반시계 방향으로 90도 이동하여 12시 방향에서 시작함
plt.title(name + ' 지역의 남녀 성별 비율')
plt.show()
```
